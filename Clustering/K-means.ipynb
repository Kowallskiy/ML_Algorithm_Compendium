```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("c:/Users/Userpc/Desktop/Projects/ML Shortcuts/Data/k-means.csv")
```

```python
sns.scatterplot(data=df, x="x", y="y")
```

```python
scaler = MinMaxScaler()
df.drop('z', axis=1, inplace=True)
df.head()
```

```python
data = np.array(scaler.fit_transform(df))
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
model = KMeans(n_clusters=5, n_init=20, max_iter=300, random_state=0)

y_pred = model.fit(data)
y_pred.labels_
```

```python
df['Cluster'] = y_pred.labels_
```

```python
df.head()
```

```python
data = np.array(df)
```

```python
x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1

sample_size = 1000 
sampled_data = data[np.random.choice(data.shape[0], sample_size, replace=False)]

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.scatter(sampled_data[:, 0], sampled_data[:, 1], c=df.Cluster.values, cmap='viridis', label='Data Points')
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.8, levels=np.arange(0, 6) - 0.5)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.legend()
plt.show()
```
